A **vector** is an ordered list of numbers.  
For example, a three-dimensional vector can be written as  

$$
\vec{v} = [2,\;4,\;-1]
$$

Each element represents a component along an axis (x, y, z).  
In Python, we use **NumPy arrays** to represent vectors efficiently.

Although you can use lists, [**NumPy**](https://https://numpy.org/)  arrays are much better because they
support fast mathematical operations.

```python
import numpy as np #call the numpy library and define it as np
v1 = np.array([2, 4, -1])
v2 = np.array([1, 3, 5])

print("Vector v1:", v1)
print("Vector v2:", v2)
```

The elements of the arrays created in the previous example can be added (addition)

```python
sum_vector = v1 + v2
print("v1 + v2 =", sum_vector)
```

Subtracted

```python
diff_vector = v1 - v2
print("v1 - v2 =", diff_vector)
```

# Plotting with Matplotlib


**Matplotlib** is Python’s most popular plotting library.  
It allows you to create line plots, scatter plots, bar charts, histograms, and even 3D graphics.

In Google Colab, Matplotlib is already installed — you just need to import it.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Let’s start with a simple example: plotting a line defined by \( y = x^2 \).

```python
x = np.linspace(-10, 10, 100)

# Compute corresponding y values
y = x ** 2

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y = x²')
plt.grid(True)
plt.title('Simple Quadratic Function')
```

`np.linspace() `creates 100 evenly spaced points from −10 to 10.
`plt.plot()` draws the line, and `plt.show()` displays the figure.

You can control color, line style, and markers using simple arguments.

```python
plt.plot(x, y, color='red', linestyle='--', marker='o', label='y = x²')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Customized Line Plot')
plt.legend()
plt.grid(True)
plt.show()
```

`color='red'` → red line

`linestyle='--' `→ dashed line

`marker='o'` → dots on data points

`plt.legend()` adds a legend, and `plt.grid(True)` adds gridlines.

We can plot several functions on the same figure to compare them.

```python
y1 = x ** 2
y2 = x ** 3
y3 = 400*np.sin(250*x)

plt.plot(x, y1, label='x²')
plt.plot(x, y2, label='x³')
plt.plot(x, y3, label='400*sin(250*x)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Multiple Curves')
plt.legend()
plt.grid(True)
plt.show()
```

Scatter plots show discrete data points instead of continuous lines.


```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='green', marker='x')
plt.title('Random Scatter Plot')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid(True)
plt.show()
```

Bar charts are great for comparing categorical data.

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 23, 7, 15]

plt.bar(categories, values, color='blue')
plt.title('Example Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

Use subplots to display multiple figures in one window.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title('Sine Function')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title('Cosine Function')
plt.grid(True)

plt.tight_layout()
plt.show()
```

The code down below shows how plot a graph from a pre-defined function

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the linear function y = ax + c
a = 2
c = 1

# Define the linear function
def linear_function(x, a, c):
  return a * x + c


# Generate x values
x = np.linspace(0, 10, 100)

# Calculate y values using the function
y = linear_function(x, a, c)

# Plot the linear function
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y = ax + c')
plt.title(f'Plot of y = {a}x + {c}')
plt.grid(True)
plt.show()
```

# Working with DataFrames

A **DataFrame** is like a table (similar to an Excel sheet).  
It consists of rows and columns, where each column can hold a different type of data (numbers, strings, etc.).

You can think of it as a powerful, flexible spreadsheet inside Python.

In python you can create a DataFrame from lists and specify the column names.

```python
import pandas as pd
data = [
    ['Chikhi', 16.75, 'V.good'],
    ['Benammour', 17.5, 'Excellent'],
    ['Elaouaber', 15, 'Good']
]

df = pd.DataFrame(data, columns=['Name', 'Score', 'Mention'])
print(df)
```
